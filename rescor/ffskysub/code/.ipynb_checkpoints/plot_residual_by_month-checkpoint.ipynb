```python
import numpy as np
import h5py
from astropy.table import Table
import glob
import os

from hetdex_api.survey import Survey

from hetdex_api.shot import open_shot_file, get_fibers_table
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors
%matplotlib inline
```

```python
filename = "../../hetdex_lim_measurement/data/ifu_spec_mw_boxcar_aperturemask_no2017_spring.h5"
tab = {}
with h5py.File(filename, 'r') as ff:
    print(ff.keys())
    shotids = ff['shotid'][:]
    ifu_here_finite = (shotids < 20230500000) * (shotids > 20230400000)
    for key in ['multiframe', 'shotid', 'biweight_spec']:
        tab[key] = ff[key][:][ifu_here_finite]
        print(key)
tab = Table(tab)
```

```python
tab = tab[np.argsort(tab['shotid'])]
multiframes = np.unique(tab['multiframe'])
```

```python
cmap = plt.cm.viridis
norm = matplotlib.colors.Normalize(vmin=min(tab['shotid']), vmax=max(tab['shotid']))
```

```python
def_wave = np.arange(3470, 5542, 2)
def_wave[134], def_wave[182]
```

```python
survey = Survey("hdr4")
survey = survey.return_astropy_table()
print(survey.columns)
```

```python
x_axis = "shotid"

for multi in multiframes[40:50]:
    xs = []
    here = tab["multiframe"] == multi
    plt.figure()#figsize=(15,3))
    for i in range(len(tab[here])):
        shot_here = survey['shotid'] == tab[here]['shotid'][i]
        xs.append((survey[x_axis][shot_here]))
        plt.plot((survey[x_axis][shot_here]), tab[here]['biweight_spec'][i][134], color='k', alpha=0.5, marker='.')
        plt.plot((survey[x_axis][shot_here]), tab[here]['biweight_spec'][i][182], color='r', alpha=0.5, marker='.')
    plt.title(multi)
    plt.xlabel(x_axis)

    try:
        xs = np.sort(xs)
        p = np.polyfit(xs, tab[here]['biweight_spec'][:,134], 2)
        plt.plot(xs, p[0] * xs**2 + p[1] * xs + p[2], color='k')
        p = np.polyfit(xs, tab[here]['biweight_spec'][:,182], 2)
        plt.plot(xs, p[0] * xs**2 + p[1] * xs + p[2], color='r')
    except:
        pass

    line, = plt.plot([], [], color='k', alpha=0.5, marker='.')
    line2, = plt.plot([], [], color='r', alpha=0.5, marker='.')
    line.set_label(def_wave[134])
    line2.set_label(def_wave[182])
    plt.legend()
```
